```python
from scipy.stats import skew
import numpy as np 
import matplotlib.pyplot as plt
```

```python
q9_a = pd.read_csv("Q9_a.csv")
```

```python
q9_a.head()
```

## Skewness and Kurtosis

```python
x1 = q9_a.skew()
x1
```

```python
x2 = q9_a.kurtosis()
x2
```

```python
plt.plot(x1,x2)
plt.xlabel('Skewness')
plt.ylabel('Kurtosis')
plt.title('Skewness and Kurtosis', size=20)
```
